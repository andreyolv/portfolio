```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/postgres')
```

```python
teste = pd.read_sql('select * from playground', con=engine)
teste
```

```python
teste.loc[-1] = [4, 5, 'teste2', 'rede', 'norte','2020-10-10']
teste
```

```python
teste.index = teste.index + 1
teste = teste.sort_index()
teste
```

```python
teste.to_sql('playground', con=engine, if_exists='replace',index=True)
```
