# Reading Table

```python
anp.sheets[0].range('C49').value = estados[15]
anp.sheets[0].range('C50').value = produtos1[5]
```

```python
# anp.sheets[0].range('B54:W65').options(ndim=2).value
df = anp.sheets[0].range('B53:W65').options(pd.DataFrame, header=1, index=False).value
```

```python
df
```

```python
df2 = pd.melt(df, id_vars='Dados', var_name='ano', value_name='volume')
df2
```

```python
df2 = df2.astype({"ano":int})
df2['ano'] = df2['ano'].round(0)
df2
```

```python
month_map = {'Janeiro':'01', 'Fevereiro':'02', 'Março':'03', 'Abril': '04', 'Maio':'05', 'Junho':'06',
          'Julho':'07', 'Agosto':'08', 'Setembro':'09', 'Outubro':'10', 'Novembro':'11', 'Dezembro':'12'}

df2['month_number'] = df2['Dados'].apply(lambda x:month_map[x])

df2.tail(20)
```

```python
df2 = df2.astype({"ano":str, "month_number":str})
df2.dtypes
```

```python
df2['year_month'] = df2.ano.str.cat(df2.month_number,sep="-")
df2
```

```python
df2 = df2.drop(['Dados', 'ano', 'month_number'], axis=1)
df2
```

```python
df2['uf'] = estados[15] 
df2['product'] = produtos1[5][0:-5]
df2['unit'] = "m3"
df2['created_at'] = pd.Timestamp.now()
df2
```

```python
column_names_ordered = ["year_month", "uf", "product", "unit", "volume", "created_at"]
df2 = df2[column_names_ordered]
df2
```

```python
df2 = df2.dropna(subset = ['volume'])
df2
```
